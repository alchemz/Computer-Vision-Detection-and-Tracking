```python
#split each image to 3 square, 720*240 --> 240*240
import os
from PIL import Image

def splitImage(src, rownum, colnum, dstpath):
    img = Image.open(src)
    w,h = img.size
    if rownum <= h and colnum <= w:
        print('Original image info: %sx%s, %s, %s' % (w, h, img.format, img.mode))
        s = os.path.split(src)
        if dstpath == '':
            dstpath = s[0]
        fn = s[1].split('.')
        basename = fn[0]
        ext = fn[-1]
        
        num = 0
        rowheight = h // rownum
        colwidth = w //colnum
        
        for r in range(rownum):
            for c in range(colnum):
                box = (c * colwidth, r * rowheight, (c + 1) * colwidth, (r + 1) * rowheight)
                img.crop(box).save(os.path.join(dstpath, basename + '_' + str(num) + '.' + ext), ext)
                num = num + 1
        print('Splitting completed，generated %s subimages。' % num)
    else:
        print('invalid parameters for cropping！')

#create folder
def mkdir(path):
    path = path.strip()
    path = path.rstrip("\\")
    
    isExists = os.path.exists(path)
    
    if not isExists:
        os.makedirs(path)
        #print(path+' create directory successfully')
        return True
    else:
        print(path+ ' directory existed')
        return False


folder = r'./Desktop/data_cropped'
path = os.listdir(folder)

for each_jpg in path:
    first_name, second_name = os.path.splitext(each_jpg)
    each_jpg = os.path.join(folder, each_jpg)
    src = each_jpg
    #remove .DS_Store hidden file
    current_file = os.path.basename(each_jpg)
    if(current_file == '.DS_Store'):
        print('a DS_Store item encountered')
        os.remove(each_jpg)

    #print(src)
    #print(first_name)
    #create directory
    mkpath = "./Desktop/data_splitted/" + first_name
    #call function
    mkdir(mkpath)
    
    if(os.path.isfile(src)):
        dstpath = mkpath
        #if dstpath == '.DS_Store':
            #print('a DS_Store item encountered')
            #os.remove(each_jpg)
        if(dstpath == '') or os.path.exists(dstpath):
            row = int(1) #rows to split
            col = int(3) #cols to split
            if row>0 and col>0:
                splitImage(src, row, col, dstpath)
            else:
                print('Invalid parameters for splitting')
        else:
            print('Directory %s not existed' % dstpath)
    else:
        print('Image %s not existed' % src)
```
